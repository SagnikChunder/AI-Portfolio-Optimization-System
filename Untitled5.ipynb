```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
data = yf.download(tickers, start='2018-01-01', end='2024-12-31')
data.dropna(inplace=True)
```

```python
data.plot(figsize=(12,6))
plt.title("Historical Adjusted Closing Prices")
plt.grid()
plt.show()
```

Return Calculation and Risk metrics

```python
daily_returns = data.pct_change().dropna()
annual_returns = daily_returns.mean() * 252
```

```python
#Volatility and Correlation
annual_volatility = daily_returns.std() * np.sqrt(252)
correlation_matrix = daily_returns.corr()
```

```python
#Sharpe Ratio
risk_free_rate = 0.04  # assume 4%
sharpe_ratios = (annual_returns - risk_free_rate) / annual_volatility
```

Optimization

```python
# Required imports (if not already imported)
import numpy as np
import pandas as pd

# --- Data Preparation (Corrected) ---
# This part is crucial to fix the error.
# We assume 'daily_returns' is a DataFrame with one column per ticker.
# And 'tickers' is a list of the ticker symbols.

print("Preparing inputs for simulation...")

# Select only 'Close' price returns from daily_returns
daily_returns_close = daily_returns['Close']

# Calculate annualized mean returns as a NumPy array of shape (5,)
mean_returns_vector = daily_returns_close.mean().to_numpy() * 252

# Calculate annualized covariance matrix as a NumPy array of shape (5, 5)
cov_matrix_np = daily_returns_close.cov().to_numpy() * 252

# --- Sanity Check (Optional but Recommended) ---
print(f"Shape of mean_returns_vector: {mean_returns_vector.shape}") # Expected: (5,)
print(f"Shape of cov_matrix_np: {cov_matrix_np.shape}")       # Expected: (5, 5)
print("-" * 30)

# --- Monte Carlo Simulation ---

# 1. Define Simulation Parameters
num_portfolios = 25000
num_tickers = len(tickers)
risk_free_rate = 0.04

# 2. Pre-allocate results array for efficiency
# Columns: Return, Volatility, Sharpe Ratio, w_AAPL, w_MSFT, ...
results_columns = 3 + num_tickers
results_array = np.zeros((num_portfolios, results_columns))

print(" Running Monte Carlo simulation...")

# 3. Simulate portfolios
for i in range(num_portfolios):
    # Generate random weights that sum to 1
    weights = np.random.random(num_tickers)
    weights /= np.sum(weights)

    # Calculate portfolio metrics using the correctly shaped inputs
    portfolio_return = np.dot(weights, mean_returns_vector)
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix_np, weights)))
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_volatility

    # Store results
    results_array[i, 0] = portfolio_return
    results_array[i, 1] = portfolio_volatility
    results_array[i, 2] = sharpe_ratio
    results_array[i, 3:] = weights

print(" Simulation complete.")

# 4. Convert results to a DataFrame for analysis
weight_cols = [f'w_{ticker}' for ticker in tickers]
results_df = pd.DataFrame(results_array, columns=['Return', 'Volatility', 'Sharpe Ratio'] + weight_cols)

# 5. Display the top 5 portfolios by Sharpe Ratio
print("\n Top 5 Portfolios by Sharpe Ratio:")
print(results_df.nlargest(5, 'Sharpe Ratio'))
```

```python
import matplotlib.pyplot as plt

# 1. Find the portfolio with maximum Sharpe Ratio
max_sharpe_idx = results_df['Sharpe Ratio'].idxmax()
max_sharpe_portfolio = results_df.loc[max_sharpe_idx]

# 2. Find the portfolio with minimum volatility
min_vol_idx = results_df['Volatility'].idxmin()
min_vol_portfolio = results_df.loc[min_vol_idx]

# 3. Plot Efficient Frontier
plt.figure(figsize=(12, 6))
scatter = plt.scatter(results_df['Volatility'], results_df['Return'],
                      c=results_df['Sharpe Ratio'], cmap='viridis', alpha=0.5)
plt.colorbar(scatter, label='Sharpe Ratio')
plt.xlabel('Volatility (Risk)')
plt.ylabel('Expected Return')
plt.title('Efficient Frontier with Optimal Portfolios')
plt.grid(True)

# 4. Mark the optimal portfolios
plt.scatter(max_sharpe_portfolio['Volatility'], max_sharpe_portfolio['Return'],
            color='red', marker='*', s=300, label='Max Sharpe Ratio')

plt.scatter(min_vol_portfolio['Volatility'], min_vol_portfolio['Return'],
            color='blue', marker='D', s=200, label='Min Volatility')

plt.legend()
plt.show()
```

```python
# Find the index of the max Sharpe Ratio portfolio
max_sharpe_idx = results_df['Sharpe Ratio'].idxmax()
max_sharpe_portfolio = results_df.loc[max_sharpe_idx]

# Print all metrics
print(" Best Portfolio by Sharpe Ratio:")
print(f"\nExpected Annual Return     : {max_sharpe_portfolio['Return']:.2%}")
print(f"Annual Volatility (Risk)   : {max_sharpe_portfolio['Volatility']:.2%}")
print(f"Sharpe Ratio               : {max_sharpe_portfolio['Sharpe Ratio']:.4f}")

print("\n Portfolio Weights:")
for ticker in tickers:
    weight = max_sharpe_portfolio[f'w_{ticker}']
    print(f"{ticker:<6}: {weight:.2%}")
```
